# 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Rating_cat']=df['Rating'].apply(rating)
```

# 22. Find Unique Values From Genre 

```python
df['Genre']
```

```python
list=[]
for value in df['Genre']:
    list.append(value.split(','))
```

```python
list
```

```python
one_d = []
for item in list:
    for item in item:
        one_d.append(item)
```

```python
one_d
```

```python
uni_list = []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
